```python
def visualize_images(model, X, num_images=10, image_shape=(28, 28)):
    """
    Visualizes the original and reconstructed images using the provided Autoencoder model.
    
    Arguments:
    model -- Trained Autoencoder model to use for reconstruction.
    X -- Input images to reconstruct.
    num_images -- (Optional) Number of images to visualize. Default is 10.
    image_shape -- (Optional) Shape of the input images. Default is (28, 28).
    
    """
    
    # Get the reconstructed images
    reconstructed_images = model.predict(X)

    # Define the subplot grid
    plt.figure(figsize =(20, 4))
    rows = 2
    columns = num_images
    
    # Loop through the images and plot them
    for i in range(num_images):
        
        # Original image subplot
        ax = plt.subplot(rows, columns, i + 1)
        plt.imshow(X[i].reshape(image_shape))
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False) 
                
        # Reconstructed image subplot
        ax = plt.subplot(rows, columns, i + 1 + num_images)
        plt.imshow(reconstructed_images[i].reshape(image_shape))
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

    # Show the plot
    plt.show()
```
